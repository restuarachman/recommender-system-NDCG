```python
# buat sebuah fungsi yang menerima input berupa user id (uid)
# mengambilkan semua item id (iid) dan original rating (r_ui) pada test set
def getUserRating(id):
  result = []
  for value in testset:

    if value[0] == id :
      result.append((value[1], value[2]))

  return result
```

```python
from sklearn.metrics import dcg_score
# buat fungsi yang menghitung nilai dari NDCG
# input user id (iid)
# output nilai NDCG = DCG / IDCG 

def calculateNDCG(id):
  # panggil fungsi yang mengambil transaksi dari uid pada test set
  data = getUserRating(id)
  
  originalRating = np.asarray(data, dtype=np.float64)[:,1]

  # Loop untuk semua transaksi yang diberikan pada perserintah sebelumnya
  estimatedRating = []
  for value in data:
    # cari prediksi rating (est)
    item = value[0]
    rating = value[1]
    estimatedRating.append(algo.predict(id, item, r_ui=rating, verbose=True).est) 

  # cari nilai DCG (urut berdasar estimated rating)
  dcg = dcg_score([originalRating], [estimatedRating])
  print("DCG  :", dcg)

  # cari nilai iDCG (urut berdasar original rating) 
  idcg = dcg_score([originalRating], [originalRating])
  print("iDCG :", idcg)

  # return NDCG = DCG / iDCG
  ndcg = dcg / idcg
  return ndcg
```

```python
inputId = input('Masukan id user : ')
ndcg = calculateNDCG(inputId)
print("NDCG :", ndcg)
```
